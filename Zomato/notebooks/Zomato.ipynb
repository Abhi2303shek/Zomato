# 1. Restaurant Performance Insights

### 1.1 Top restaurants by rating, votes, and delivery

```python
zomato['Average_Rating'] = pd.to_numeric(zomato['Average_Rating'], errors='coerce')
zomato['Votes'] = pd.to_numeric(zomato['Votes'], errors='coerce').fillna(0)
```

```python
agg = zomato.groupby(['Restaurant_Name', 'Place_Name'], as_index=False).agg({
    'Average_Rating':'mean',   
    'Votes':'sum',
    'Delivery_Rating':'mean',
    'Dining_Rating':'mean'
})
```

```python
best_restaurant = agg[(agg['Average_Rating']>=4.2) & (agg['Votes']>=10000)].sort_values(by=['Average_Rating', 'Votes'], ascending=[False, False]).head(10)   
best_restaurant
```

```python
best_restaurant_by_del = agg[agg['Delivery_Rating']>=4.2].sort_values(by=['Delivery_Rating', 'Votes'], ascending=[False, False]).head(10)   
best_restaurant_by_del
```

### 1.2 Restaurants with highest total votes

```python
best_restaurant_by_votes = agg[agg['Delivery_Rating']>=4.2].sort_values(by=['Votes'], ascending=[False]).head(5)   
best_restaurant_by_votes
```

### 1.3 Compare and analyze the difference between Dining_Rating and Delivery_Rating.

```python
zomato['diff'] = agg['Delivery_Rating'] - agg['Dining_Rating']
zomato['diff']
```

```python
restaurant_summary = zomato.groupby('Restaurant_Name')['diff'].mean().reset_index().sort_values(
    by='diff',
    ascending=False).head(10)
restaurant_summary
```

```python
zomato['din_diff'] = agg['Dining_Rating'] - agg['Delivery_Rating']
zomato['din_diff']
```

```python
restaurant_summary = zomato.groupby('Restaurant_Name')['din_diff'].mean().reset_index().sort_values(
    by='din_diff',
    ascending=False)
restaurant_summary
```

### 1.4 Determine the quantitative effect of the Best_Seller status.

```python
best_restaurant_bs = agg[(agg['Votes']>=10000) & (zomato['Best_Seller']=='BESTSELLER')].sort_values(by=['Votes'], ascending=[False])   
best_restaurant_bs
```

### 1.5 Analyse the characteristics of restaurants where Is_Highly_Rated is 1.

```python
common = zomato.groupby(['Restaurant_Name','Cuisine','City'])['Is_Highly_Rated'].mean().reset_index().sort_values(
    by='Is_Highly_Rated',
    ascending=False).head(10)
common
```

```python
common = zomato.groupby(['Restaurant_Name','Cuisine','Prices','City'])['Is_Highly_Rated'].mean().reset_index().sort_values(
    by='Is_Highly_Rated',
    ascending=False).head(10)
common
```

# 2. City Performance Benchmarking

### 2.1 Identify the city with the highest average restaurant rating

```python
high_avg_city = zomato.groupby('City')['Avg_Rating_City'].mean().reset_index().sort_values(
    by='Avg_Rating_City',
    ascending=False).head(10)
high_avg_city
```

### 2.2 Identify the city with the highest average food price (Avg_Price_City)

```python
high_avg_price_city = zomato.groupby('City')['Avg_Price_City'].mean().reset_index().sort_values(
    by='Avg_Price_City',
    ascending=False).head(10)
high_avg_price_city
```

### 2.3 Which localities show the highest concentration of high-rated or popular restaurants?

```python
high_restau_city = zomato.groupby('City')['Restaurant_Name'].nunique().reset_index()   
high_restau_city.sort_values(by='Restaurant_Name',ascending=False).head(10)
```

# 3. Pricing and Value Analysis

### 3.1. Analyze the relationship between an item's Average_Rating and its calculated Price_per_Vote.

```python
pricing_and_value_analysis = zomato.groupby('Item_Name')[['Price_per_Vote','Average_Rating']].mean().reset_index().sort_values(
    by='Average_Rating',
    ascending=False)
pricing_and_value_analysis
```

### 3.1. Compare the Prices of items with high Votes versus items with low Votes within the same Cuisine and Restaurant_Name.

```python
zomato['Is_Popular'] = zomato['Votes'] > zomato['Votes'].median()
```

```python
price_comparison = zomato.groupby(['Cuisine', 'Is_Popular'])['Prices'].mean().reset_index().head(10)
price_comparison
```

### 3.2. Analyze the attributes of restaurants where the Is_Expensive column is 1.

```python
Expensive_prices = zomato.groupby('Is_Expensive')['Prices'].mean().reset_index().head(10)
Expensive_prices
```

```python
Expensive_cuisine = zomato.groupby('Is_Expensive')['Avg_Price_Cuisine'].mean().reset_index().head(10)
Expensive_cuisine
```

```python
Expensive_cuisine = zomato.groupby('Is_Expensive')['Restaurant_Popularity'].mean().reset_index().head(10)
Expensive_cuisine
```

# 4. Cuisine Analysis

### 4.1. Rank all Cuisine types based on the highest total Total_Votes across all restaurants (Cuisine Popularity Ranking).

```python
most_voted_cuisine = zomato.groupby('Cuisine')['Total_Votes'].sum().reset_index().sort_values(by='Total_Votes',ascending=False).head(10)
most_voted_cuisine
```

### 4.2. Rank all Cuisine types based on their calculated Avg_Rating_Cuisine (Cuisine Quality Ranking).

```python
most_rated_cuisine = zomato.groupby('Cuisine')['Avg_Rating_Cuisine'].mean().reset_index().sort_values(by='Avg_Rating_Cuisine',ascending=False).head(10)
most_rated_cuisine
```

### 4.3. Compare the Avg_Price_Cuisine to understand which cuisine types are, on average, the most and least expensive (Cuisine Pricing Analysis).

```python
most_priced_cuisine = zomato.groupby('Cuisine')['Avg_Price_Cuisine'].mean().reset_index().sort_values(by='Avg_Price_Cuisine',ascending=False).head(10)
most_priced_cuisine
```

```python
least_priced_cuisine = zomato.groupby('Cuisine')['Avg_Price_Cuisine'].mean().reset_index().sort_values(by='Avg_Price_Cuisine',ascending=True).head(10)
least_priced_cuisine
```
